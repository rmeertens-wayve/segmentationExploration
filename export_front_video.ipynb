```python
import cv2
from pathlib import Path

source_folder = Path('/Users/rolandmeertens/datasets/driveSafeAI/v0/scene_3/rgb_undistorted')
video_name = 'scene_3_front.mp4'
fps = 30


img_array = list()
for i, filename in enumerate(sorted(source_folder.glob('*_ff.jpeg'))): 
    print(filename)
    img = cv2.imread(str(filename))

    height, width, layers = img.shape
    size = (width,height)
    img_array.append(img)

#Create Video
out = cv2.VideoWriter(video_name,cv2.VideoWriter_fourcc(*'MP4V'), fps, size)
for i in range(len(img_array)):
    out.write(img_array[i])
out.release()
```
